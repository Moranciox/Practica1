```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# Definir la funcin para visualizacin y anlisis
def visualizar_analisis_foliares(ensayo, df, pdf_pages):
    # Filtrar por ensayo
    df_ensayo = df[df['Ensayo'] == ensayo]

    # Variables numricas
    variables_numericas = ['N (%)', 'Ca (%)', 'K (%)', 'Mg (%)', 'Na (%) (<0,025)', 'P (%)', 'S (%)', 'B (ppm)', 'Cu (ppm) (<5)', 'Fe (ppm)', 'Mn (ppm)', 'Zn (ppm)']

    # Visualizacin: Boxplot comparativo entre fechas de muestreo para las variables numricas
    plt.figure(figsize=(18, 10))
    plt.suptitle(f'Anlisis Foliares - {ensayo}', fontsize=16)
    
    for i, variable in enumerate(variables_numericas, 1):
        plt.subplot(3, 4, i)
        sns.boxplot(x='Fecha muestreo', y=variable, data=df_ensayo)
        plt.title(variable)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    pdf_pages.savefig()
    plt.close()

# Ruta del archivo Excel
ruta_excel = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'

# Cargar el archivo Excel en un DataFrame
df_foliares = pd.read_excel(ruta_excel, sheet_name='Anlisis Foliares')

# Crear un archivo PDF para guardar los resultados
ruta_pdf = '/home/moranciox/Practica/Excel_practicas/Excel_Resumen/Resultados_Analisis_Foliares.pdf'
with PdfPages(ruta_pdf) as pdf:
    # Visualizar y analizar para Emiliana
    visualizar_analisis_foliares('Emiliana', df_foliares, pdf)

    # Visualizar y analizar para Cono Sur
    visualizar_analisis_foliares('Cono Sur', df_foliares, pdf)

    # Visualizar y analizar para Concha y Toro
    visualizar_analisis_foliares('Concha y Toro', df_foliares, pdf)

print(f"Resultados guardados en: {ruta_pdf}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar tus datos en un DataFrame de Pandas
ruta_archivo = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'
df = pd.read_excel(ruta_archivo, sheet_name="Anlisis Foliares")

# Filtrar datos para el ensayo Emiliana y los aos 2022 y 2023
df_emiliana_2022 = df[(df['Ensayo'] == 'Emiliana') & (df['Fecha muestreo'].dt.year == 2022)]
df_emiliana_2023 = df[(df['Ensayo'] == 'Emiliana') & (df['Fecha muestreo'].dt.year == 2023)]

# Seleccionar las columnas relevantes para el anlisis de correlacin
columnas_interes = ['N (%)', 'Ca (%)', 'K (%)', 'Mg (%)', 'Na (%) (<0,025)', 'P (%)', 'S (%)', 'B (ppm)', 'Cu (ppm) (<5)', 'Fe (ppm)', 'Mn (ppm)', 'Zn (ppm)']
df_emiliana_2022 = df_emiliana_2022[columnas_interes]
df_emiliana_2023 = df_emiliana_2023[columnas_interes]

# Calcular las matrices de correlacin para ambos aos
matriz_correlacion_2022 = df_emiliana_2022.corr()
matriz_correlacion_2023 = df_emiliana_2023.corr()

# Crear mapas de calor utilizando Seaborn
fig, axs = plt.subplots(1, 2, figsize=(18, 6))

sns.heatmap(matriz_correlacion_2022, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=axs[0])
axs[0].set_title('Matriz de Correlacin - Ao 2022')

sns.heatmap(matriz_correlacion_2023, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=axs[1])
axs[1].set_title('Matriz de Correlacin - Ao 2023')

plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# Cargar tus datos en un DataFrame de Pandas
ruta_archivo = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'
df_suelo = pd.read_excel(ruta_archivo, sheet_name="Anlisis de Suelo")

# Filtrar datos para los ensayos Emiliana, Cono Sur y Concha y Toro
ensayos_interes = ['Emiliana', 'Cono Sur', 'Concha y Toro']
df_suelo = df_suelo[df_suelo['Ensayo'].isin(ensayos_interes)]

# Columnas de inters
columnas_suelo = ['P (ppm)', 'N (ppm)', 'H2O (pH)', 'K disp. (ppm)', 'M.O (%)', 
                  'Ca (cmol (+) /kg)', 'Mg (cmol (+) /kg)', 'K (cmol (+) /kg)', 'Na (cmol (+) /kg)',
                  'Suma de bases', 'Zn (ppm)', 'Fe (ppm)', 'Cu (ppm)', 'Mn (ppm)', 'Boro (ppm)']

# Crear un archivo PDF para guardar los resultados
ruta_pdf_suelo = '/home/moranciox/Practica/Resultados_Analisis_Suelo.pdf'
pdf_pages_suelo = PdfPages(ruta_pdf_suelo)

# Boxplots y Matrices de correlacin
for anio in df_suelo['Ao de plantacin'].unique():
    df_suelo_anio = df_suelo[df_suelo['Ao de plantacin'] == anio]

    plt.figure(figsize=(18, 12))
    for i, columna in enumerate(columnas_suelo, 1):
        plt.subplot(4, 4, i)
        sns.boxplot(x='Ensayo', y=columna, data=df_suelo_anio)
        plt.title(columna)

    plt.tight_layout()
    plt.suptitle(f'Boxplots - Anlisis de Suelo - Ao {anio}', y=1.02, fontsize=16)
    pdf_pages_suelo.savefig()
    plt.close()

    for ensayo in ensayos_interes:
        df_ensayo_suelo_anio = df_suelo[(df_suelo['Ensayo'] == ensayo) & (df_suelo['Ao de plantacin'] == anio)]
        df_ensayo_suelo_anio = df_ensayo_suelo_anio[columnas_suelo]

        # Calcular la matriz de correlacin
        matriz_correlacion_suelo_anio = df_ensayo_suelo_anio.corr()

        # Crear mapa de calor utilizando Seaborn
        plt.figure(figsize=(12, 10))
        sns.heatmap(matriz_correlacion_suelo_anio, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
        plt.title(f'Matriz de Correlacin - Anlisis de Suelo - {ensayo} - Ao {anio}')

        pdf_pages_suelo.savefig()
        plt.close()

pdf_pages_suelo.close()
print(f"Resultados de suelo guardados en: {ruta_pdf_suelo}")


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.backends.backend_pdf import PdfPages

# Definir la funcin para visualizacin y anlisis
def visualizar_analisis_suelo(ensayo, df, pdf_pages):
    # Filtrar por ensayo
    df_ensayo = df[df['Ensayo'] == ensayo]

    # Variables numricas
    variables_numericas = ['P (ppm)', 'N (ppm)', 'H2O (pH)', 'K disp. (ppm)', 'M.O (%)', 
                           'Ca (cmol (+) /kg)', 'Mg (cmol (+) /kg)', 'K (cmol (+) /kg)', 'Na (cmol (+) /kg)',
                           'Suma de bases', 'Zn (ppm)', 'Fe (ppm)', 'Cu (ppm)', 'Mn (ppm)', 'Boro (ppm)']

    # Visualizacin: Boxplot comparativo entre fechas de muestreo para las variables numricas
    plt.figure(figsize=(18, 10))
    plt.suptitle(f'Anlisis de Suelo - {ensayo}', fontsize=16)
    
    for i, variable in enumerate(variables_numericas, 1):
        plt.subplot(3, 5, i)
        sns.boxplot(x='Fecha muestreo', y=variable, data=df_ensayo)
        plt.title(variable)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    pdf_pages.savefig()
    plt.close()

# Ruta del archivo Excel
ruta_excel = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'

# Cargar el archivo Excel en un DataFrame
df_suelo = pd.read_excel(ruta_excel, sheet_name='Anlisis de Suelo')

# Crear un archivo PDF para guardar los resultados
ruta_pdf_suelo = '/home/moranciox/Practica/Excel_practicas/Excel_Resumen/Resultados_Analisis_Suelo.pdf'
with PdfPages(ruta_pdf_suelo) as pdf_suelo:
    # Visualizar y analizar para Emiliana
    visualizar_analisis_suelo('Emiliana', df_suelo, pdf_suelo)

    # Visualizar y analizar para Cono Sur
    visualizar_analisis_suelo('Cono Sur', df_suelo, pdf_suelo)

    # Visualizar y analizar para Concha y Toro
    visualizar_analisis_suelo('Concha y Toro', df_suelo, pdf_suelo)

print(f"Resultados de Anlisis de Suelo guardados en: {ruta_pdf_suelo}")
```

```python
import pandas as pd

# Cargar los datos
ruta_archivo = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'
df = pd.read_excel(ruta_archivo, sheet_name="Anlisis de Suelo")

# Filtrar datos para Emiliana, Cono Sur y Concha y Toro
ensayos_seleccionados = ['Emiliana', 'Cono Sur', 'Concha y Toro']
df_ensayos = df[df['Ensayo'].isin(ensayos_seleccionados)]

# Filtrar datos por aos (2022 y 2023)
df_2022 = df_ensayos[df_ensayos['Fecha muestreo'].dt.year == 2022]
df_2023 = df_ensayos[df_ensayos['Fecha muestreo'].dt.year == 2023]

# Seleccionar las columnas de inters
columnas_interes = ['P (ppm)', 'N (ppm)', 'H2O (pH)', 'K disp. (ppm)', 'M.O (%)', 
                    'Ca (cmol (+) /kg)', 'Mg (cmol (+) /kg)', 'K (cmol (+) /kg)', 'Na (cmol (+) /kg)',
                    'Suma de bases', 'Zn (ppm)', 'Fe (ppm)', 'Cu (ppm)', 'Mn (ppm)', 'Boro (ppm)']

# Calcular estadsticas descriptivas para 2022
estadisticas_2022 = df_2022.groupby(['Ensayo'])[columnas_interes].describe()

# Calcular estadsticas descriptivas para 2023
estadisticas_2023 = df_2023.groupby(['Ensayo'])[columnas_interes].describe()

# Mostrar los resultados en el entorno de Jupyter
print("Estadsticas Descriptivas para 2022:")
display(estadisticas_2022)

print("\nEstadsticas Descriptivas para 2023:")
display(estadisticas_2023)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta del archivo Excel
ruta_excel = '/home/moranciox/Practica/Excel_practicas/Originales/Matriz anlisis Temporada 2021-2022.xlsx'

# Cargar el archivo Excel en un DataFrame
df_suelo = pd.read_excel(ruta_excel, sheet_name='Anlisis de Suelo')

# Filtrar por ensayo 'Cono Sur'
df_cono_sur_suelo = df_suelo[df_suelo['Ensayo'] == 'Cono Sur']

# Eliminar "Sobrehilera" y "Entrehilera" de los nombres de los tratamientos
df_cono_sur_suelo['Tratamiento'] = df_cono_sur_suelo['Tratamiento'].str.replace('(Sobrehilera|Entrehilera)', '', regex=True)

# Filtrar solo las columnas necesarias
columns_of_interest_suelo = ['Tratamiento', 'Fecha muestreo', 'P (ppm)', 'N (ppm)', 'K disp. (ppm)', 'Zn (ppm)', 'Fe (ppm)', 'Cu (ppm)', 'Mn (ppm)', 'Boro (ppm)', 'S (ppm)']
df_filtered_suelo = df_cono_sur_suelo[columns_of_interest_suelo]

# Calcular lmites dinmicos con margen para todos los datos
margin_suelo = 0.2

# Obtener la lista de aos ordenada
sorted_years_suelo = sorted(df_filtered_suelo['Fecha muestreo'].dt.year.unique())

# Crear los boxplots por ao con lmites del eje Y dinmicos y margen consistente para cada columna
for column_suelo in ['P (ppm)', 'N (ppm)', 'K disp. (ppm)', 'Zn (ppm)', 'Fe (ppm)', 'Cu (ppm)', 'Mn (ppm)', 'Boro (ppm)', 'S (ppm)']:
    plt.figure(figsize=(18, 12))
    
    for i, year_suelo in enumerate(sorted_years_suelo, 1):
        plt.subplot(1, 3, i)
        df_year_suelo = df_filtered_suelo[df_filtered_suelo['Fecha muestreo'].dt.year == year_suelo]

        sns.boxplot(x='Tratamiento', y=column_suelo, data=df_year_suelo).set_title(f'Ao {year_suelo}')
        plt.ylim(df_filtered_suelo[column_suelo].min() - margin_suelo, df_filtered_suelo[column_suelo].max() + margin_suelo)  # Establecer lmites del eje Y dinmicamente con margen consistente

    # Configurar la leyenda y el ttulo
    plt.suptitle(f'Boxplots de {column_suelo} - Cono Sur', y=1.02)
    plt.tight_layout()

    # Guardar la figura en un archivo png
    plt.savefig(f'boxplots_cono_sur_{column_suelo.lower().replace(" ", "_")}_por_ao_con_margen_consistente_ordenado.png')

    # Mostrar la figura
    plt.show()

```
